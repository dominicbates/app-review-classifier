```python
from collections import Counter

# Get split of classifiers (names)
name_dist = pd.DataFrame(Counter(cleaned_df['Your Name']).most_common(),columns=['name','count'])

# Get split of categories
category_dist = pd.DataFrame(Counter(cleaned_df['Category'][pd.notnull(cleaned_df['Category'])]).most_common(),columns=['category','count'])

# Get split of semtiments
sentiment_dist = pd.DataFrame(Counter(cleaned_df['Sentiment'][pd.notnull(cleaned_df['Sentiment'])]).most_common(),columns=['sentiment','count'])

# Get split of ratings
rating_dist = pd.DataFrame(Counter(cleaned_df['Rating'][pd.notnull(cleaned_df['Rating'])]).most_common(),columns=['rating','count'])


```

```python
def make_pie(dist, rot = False):
    labels = dist.iloc[:,0]
    sizes = dist.iloc[:,1]
    title = dist.columns[0]
    
    fig1, ax1 = plt.subplots()
    ax1.pie(sizes, labels=labels, autopct='%1.1f%%',
            shadow=False, startangle=90,rotatelabels = rot)
    
count=0
for dist in [name_dist, category_dist, sentiment_dist, rating_dist]:
    if count==0:
        make_pie(dist, True)
    else:
        make_pie(dist)
    count+=1
```

### Fit some models

```python
    
    
# def get_cleaner_embedder(df, lemmatize_tf=True, w2v_vector_size=50, w2v_window_size=5):
    
#     # Set config
#     word2vec_config = {'word2vec':{'min_count':5, # Min number of occurances of word to consider
#                                    'vector_size':w2v_vector_size, # Dimension of embedding matrix (i.e. number of embedding features)
#                                    'window':w2v_window_size, # Size of window to consider
#                                    'workers':4, # Number of partitions during training
#                                    'sg':1}}


#     # Make cleander
#     cleaner = processing.TextCleaner(config = {'stop_words':True,
#                                                'ngrams':True,
#                                                'lemmatization':lemmatize_tf})

#     # Train text_cleaner on full sample
#     cleaned_text_all = cleaner.process_raw_text(df['Review'],train_ngrams = True)

#     # Train embedders on full sample
#     embedder = processing.Embedder(method='word2vec')
#     embedder.fit(cleaned_text_all)
    
#     return cleaner, embedder

def get_cleaner(df, lemmatize_tf=True):

    # Make cleander
    cleaner = processing.TextCleaner(config = {'stop_words':True,
                                               'ngrams':True,
                                               'lemmatization':lemmatize_tf})

    # Train text_cleaner on full sample
    cleaned_text_all = cleaner.process_raw_text(df['Review'],train_ngrams = True)

    return cleaner, cleaned_text_all


def get_embedder(cleaned_text_all, w2v_vector_size=50, w2v_window_size=5):
    
    # Set config
    word2vec_config = {'min_count':5, # Min number of occurances of word to consider
                       'vector_size':w2v_vector_size, # Dimension of embedding matrix (i.e. number of embedding features)
                       'window':w2v_window_size, # Size of window to consider
                       'workers':4, # Number of partitions during training
                       'sg':1}

    # Train embedders on full sample
    embedder = processing.Embedder(method='word2vec', config = word2vec_config)
    embedder.fit(cleaned_text_all)
    
    return embedder

def get_embedder_tfidf(cleaned_text_all):
    
    # Set config
    tfidf_config = {'min_df':5, # Min number of occurances of word to consider
                            'ngram_range':(1, 1)} #Â Ngram range

    # Train embedders on full sample
    embedder = processing.Embedder(method='tfidf', config = tfidf_config)
    embedder.fit(cleaned_text_all)
    
    return embedder

def process_all_samples(cleaner, embedder, df_train, df_validate, df_test):
    train_X = embedder.apply(cleaner.process_raw_text(df_train['Review']))
    validate_X = embedder.apply(cleaner.process_raw_text(df_validate['Review']))
    test_X = embedder.apply(cleaner.process_raw_text(df_test['Review']))
    return train_X, validate_X, test_X

```

```python

t1=time.time()

df_train, df_validate, df_test = split_sample(cleaned_df, 0.7, 0.2) # 0.1



model_perf_train = {} 
model_perf_validate = {} 



# Should we lemmatize text first?
for lemmatize_tf in [False]:#[False, True]:
    
    cleaner, cleaned_text_all = get_cleaner(df, lemmatize_tf=lemmatize_tf) # Get cleaner
    
    # What word2vec params should we pick?
    for w2v_vector_size in [50]:#[25,50,100]:
        for w2v_window_size in [5]:#[5, 10]:
            
            
            embedder = get_embedder(cleaned_text_all, w2v_vector_size=w2v_vector_size, w2v_window_size=w2v_window_size) # Get embedder
                
            # Apply cleaner and embedder to get samples
            X_train, X_validate, X_test = process_all_samples(cleaner, embedder, df_train, df_validate, df_test)
            y_train = df_train['Category'].values
            y_validate = df_validate['Category'].values
            y_test = df_test['Category'].values
            
            
            for xgb_max_depth in [6]:
                for xgb_eta in [0.3]:
                    for xgb_min_child_weight in [1,3]:
                        for xgb_lambda in [1,3]:
                            for xgb_subsample in [1,0.5]:
                                for xgb_num_parallel_tree in [10,1]:
                    
                
                                    model_name = 'model_'+str(lemmatize_tf)+'_'+str(w2v_vector_size)+'_'+str(w2v_window_size)+'_'+str(xgb_max_depth)+'_'+str(xgb_eta)+'_'+str(xgb_min_child_weight)+'_'+str(xgb_lambda)+'_'+str(xgb_subsample)+'_'+str(xgb_num_parallel_tree)
                                    print('\n\n********************************\nModel name: '+model_name+'\n********************************\n')


                                    # Train model
                                    model = Classifier(classifier='XGBoost', config={'max_depth':xgb_max_depth,
                                                                                     'eta':xgb_eta,
                                                                                     'min_child_weight':xgb_min_child_weight,
                                                                                     'lambda':xgb_lambda,
                                                                                     'subsample':xgb_subsample,
                                                                                     'num_parallel_tree':xgb_num_parallel_tree,})
                                    model.fit(X_train, y_train)

                                    #Â Get stats
                                    model_perf_train[model_name] = model.print_performance(y_train, model.apply(X_train), print_output=False), updated_accuracy(model.print_performance(y_train, model.apply(X_train), print_output=False))
                                    model_perf_validate[model_name] = model.print_performance(y_validate, model.apply(X_validate), print_output=False), updated_accuracy(model.print_performance(y_validate, model.apply(X_validate), print_output=False))




            

print('DONE IN:',(time.time()-t1)/60,'minutes!')
```

```python
# model_perf_validate

perf_df = []
for model_name in model_perf_validate:
    perf_df.append( pd.DataFrame({'name':model_name,'accuracy':model_perf_validate[model_name][1]}, index=[0]))
    
perf_df = pd.concat(perf_df, ignore_index=True).sort_values('accuracy', ascending=False)
```
