```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

###Â Load data

```python
df = pd.read_csv('/Users/dominicbates/Documents/GitHub/app-review-classifier/data/production_output/first_output.csv')
```

```python
df['date_dt'] = pd.to_datetime(df['date'])
```

```python
plt.figure(figsize=(9,4))
for c in set(df['category']):
    y = df[df['category']==c].groupby(pd.Grouper(key = 'date_dt', freq='M')).category.count()
    plt.plot(y,label=c)
plt.legend()
```

```python
plt.figure(figsize=(9,4))
tot = df.groupby(pd.Grouper(key = 'date_dt', freq='M')).sentiment.count()
for c in set(df['category']):
    y = df[df['category']==c].groupby(pd.Grouper(key = 'date_dt', freq='M')).category.count()
    plt.plot(y/tot,label=c)
plt.legend()
plt.ylim([None,0.65])
plt.xlabel('Month')
plt.ylabel('Category Split')
```

```python
plt.figure(figsize=(9,4))
for c in set(df['sentiment']):
    y = df[df['sentiment']==c].groupby(pd.Grouper(key = 'date_dt', freq='M')).sentiment.count()
    plt.plot(y,label=c)
plt.legend()
```

```python
plt.figure(figsize=(9,4))
tot = df.groupby(pd.Grouper(key = 'date_dt', freq='M')).sentiment.count()
for c in set(df['sentiment']):
    y = df[df['sentiment']==c].groupby(pd.Grouper(key = 'date_dt', freq='M')).sentiment.count()
    plt.plot(y/tot,label=c)
plt.legend()
plt.xlabel('Month')
plt.ylabel('Sentiment Split')
```
